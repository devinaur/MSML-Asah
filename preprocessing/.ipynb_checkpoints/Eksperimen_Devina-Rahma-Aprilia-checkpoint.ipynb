# **1. Perkenalan Dataset**

1. **Sumber Dataset**:  
   Dataset yang digunakan pada eksperimen ini adalah Mobile Phone Price Classification Dataset yang diperoleh dari platform Kaggle. Dataset ini dirancang untuk membantu memahami hubungan antara spesifikasi teknis sebuah ponsel dengan rentang harga jual (price range).

    Dataset ini umum digunakan untuk classification, di mana model akan mengelompokkan harga ke dalam beberapa kategori tertentu. Dengan karakteristik tersebut, dataset ini sangat sesuai untuk pengembangan dan evaluasi model machine learning berbasis klasifikasi.

    Dataset terdiri dari 2.000 sampel data ponsel dengan berbagai atribut spesifikasi perangkat keras dan fitur pendukung lainnya, serta satu variabel target berupa price_range. Dengan semua fitur yang tersedia, kita dapat menganalisis pengaruh spesifikasi ponsel terhadap rentang harga jualnya.

2. **Fitur Utama**:
    - battery_power: Kapasitas baterai ponsel (dalam mAh), menunjukkan daya tahan perangkat.
    - blue: Menunjukkan apakah ponsel mendukung Bluetooth.
    - clock_speed: Kecepatan prosesor ponsel (dalam GHz).
    - dual_sim: Menunjukkan apakah ponsel mendukung dual SIM.
    - fc: Resolusi kamera depan (dalam megapiksel).
    - four_g: Menunjukkan apakah ponsel mendukung jaringan 4G.
    - int_memory: Kapasitas memori internal (dalam GB).
    - m_dep: Ketebalan ponsel (dalam cm).
    - mobile_wt: Berat ponsel (dalam gram).
    - n_cores: Jumlah core prosesor.
    - pc: Resolusi kamera utama (dalam megapiksel).
    - px_height: Resolusi tinggi layar dalam piksel
    - px_width: Resolusi lebar layar dalam piksel.
    - ram: Kapasitas RAM ponsel (dalam MB).
    - sc_h: Tinggi layar ponsel (dalam cm).
    - sc_w: Lebar layar ponsel (dalam cm).
    - talk_time: Lama waktu telepon aktif (dalam jam).
    - three_g: Menunjukkan apakah ponsel mendukung jaringan 3G.
    - touch_screen: Menunjukkan apakah ponsel memiliki layar sentuh.
    - wifi: Menunjukkan apakah ponsel mendukung WiFi.

# **5. Data Preprocessing**

Kesimpulan Data Preprocessing 
- Feature engineering untuk pixel height and width
- Memisahkan kolom target, memisahkan kolom numerik dari kolom biner
- Melakukan scaling untuk kolom numerik
- Menyimpan dataset hasil preprocessing

```python
#saya melakukan feature engineering untuk mendapatkan total resolusi
df['pixel_total'] = df['px_height'] * df['px_width']
```

```python
#memisahkan kolom target
X = df.drop(columns=['price_range'])
y = df['price_range']
```

```python
#persiapan untuk scaling fitur numerik
binary_features = [
    'blue', 'dual_sim', 'four_g',
    'three_g', 'touch_screen', 'wifi'
]

numeric_features = [
    'battery_power', 'clock_speed', 'fc', 'int_memory',
    'm_dep', 'mobile_wt', 'n_cores', 'pc',
    'px_height', 'px_width', 'ram',
    'sc_h', 'sc_w', 'talk_time', 'pixel_total'
]
```

```python
#scaling
scaler = StandardScaler()

X[numeric_features] = scaler.fit_transform(X[numeric_features])
```

```python
#menyimpan dataset hasil preprocessing
processed_df = X.copy()
processed_df['price_range'] = y

processed_df.to_csv(
    "../preprocessing/phoneprice_preprocessing.csv",
    index=False
)
```
